```python
def show_available_pokemon():
    """Displays all available Pok√©mon."""
    print("\nAvailable Pok√©mon: " + ", ".join(pokemon_data.keys()))

def calculate_effectiveness(move_type, opponent_type):
    if opponent_type in type_chart[move_type]["strong"]:
        return 1.5
    elif opponent_type in type_chart[move_type]["weak"]:
        return 0.75
    return 1.0

def choose_pokemon(player):
    """Prompts the player to choose a Pok√©mon."""
    while True:
        choice = input(f"\n{player}, choose your Pok√©mon: ").strip().capitalize()
        if choice in pokemon_data:
            return choice
        print("Invalid choice, try again.")


def switch_pokemon(player, team, current_pokemon):
    """Allows player to switch Pok√©mon."""
    while True:
        print("\nYour team:")
        for i, pkm in enumerate(team, start=1):
            if pkm["name"] != current_pokemon["name"]:
                print(f"{i}. {pkm['name']} ({pkm['HP']} HP, {pkm['Energy']} Energy)")
        try:
            choice = int(input("Choose a Pok√©mon to switch to: ")) - 1
            if 0 <= choice < len(team) and team[choice]["name"] != current_pokemon["name"]:
                print(f"\n{player} switched to {team[choice]['name']}!")
                return team[choice]
            print("Invalid choice, try again.")
        except ValueError:
            print("Invalid input, enter a number.")

def apply_weather_effects(weather, p1, p2):
    if weather in ["Sandstorm", "Hail"]:
        if p1["Type"] not in weather_conditions[weather]["damage_types"]:
            p1["HP"] -= weather_conditions[weather]["damage"]
            print(f"{p1['name']} took {weather_conditions[weather]['damage']} damage from the {weather}!")
        if p2["Type"] not in weather_conditions[weather]["damage_types"]:
            p2["HP"] -= weather_conditions[weather]["damage"]
            print(f"{p2['name']} took {weather_conditions[weather]['damage']} damage from the {weather}!")

def display_status(pokemon):
    """Shows the Pok√©mon's HP, Energy, and available moves."""
    print(f"\n{pokemon['name']}'s Status: {pokemon['HP']} HP, {pokemon['Energy']} Energy")
    print(f"{pokemon['name']}'s Moves:")
    for i, (move, data) in enumerate(pokemon["Moves"].items(), start=1):
        print(f"  {i}. {move} (Power: {data['Power']}, Type: {data['Type']}, Accuracy: {data['Accuracy']}, Energy: {data['Energy Cost']})")

def apply_status_effect(pokemon):
    """Handles ongoing status effects like Poison, Burn, Sleep, etc."""
    if pokemon["Status"] == "Paralysis" and random.random() < status_effects["Paralysis"]["skip_chance"]:
        print(f"{pokemon['name']} is paralyzed and couldn't move!")
        return False

    if pokemon["Status"] == "Sleep":
        if pokemon["turns"] > 0:
            print(f"{pokemon['name']} is still asleep!")
            pokemon["turns"] -= 1
            return False
        else:
            print(f"{pokemon['name']} woke up!")
            pokemon["Status"] = None

    if pokemon["Status"] == "Burn":
        burn_damage = math.ceil(pokemon_data[pokemon["name"]]["HP"] * status_effects["Burn"]["damage_percent"])
        pokemon["HP"] -= burn_damage
        print(f"{pokemon['name']} is hurt by its burn and loses {burn_damage} HP!")

    if pokemon["Status"] == "Poison":
        pokemon["HP"] -= status_effects["Poison"]["damage"]
        print(f"{pokemon['name']} is hurt by poison and loses {status_effects['Poison']['damage']} HP!")

    if pokemon["Status"] == "Freeze" and random.random() < status_effects["Freeze"]["thaw_chance"]:
        print(f"{pokemon['name']} thawed out!")
        pokemon["Status"] = None
    elif pokemon["Status"] == "Freeze":
        print(f"{pokemon['name']} is frozen and can't move!")
        return False

    return True

def execute_move(pokemon, opponent, move, weather):
    """Executes a move, applies damage, status effects, and weather effects."""
    move_info = pokemon["Moves"][move]

    if random.randint(1, 100) > move_info["Accuracy"]:
        print(f"{pokemon['name']} used {move}, but it missed!")
        return

    if move_info["Energy Cost"] > pokemon["Energy"]:
        print(f"{pokemon['name']} doesn't have enough energy for {move}!")
        return False

    pokemon["Energy"] -= move_info["Energy Cost"]
    damage = move_info["Power"] + random.randint(1, 10)

    if random.random() < 0.1:
        damage *= 2
        print("üî• Critical Hit! The attack dealt double damage!")

    effectiveness = calculate_effectiveness(move_info["Type"], opponent["Type"])
    if effectiveness > 1:
        print("It's super effective!")
    elif effectiveness < 1:
        print("It's not very effective...")
    damage = math.ceil(damage * effectiveness)

    if weather_conditions[weather].get("boost") == move_info["Type"]:
        damage *= 1.5
        print("The weather boosted the attack!")

    opponent["HP"] -= damage
    print(f"{pokemon['name']} used {move} and dealt {damage} damage!")

    if "Effect" in move_info:
        opponent["Status"] = move_info["Effect"]
        opponent["Status Duration"] = 2
        print(f"{opponent['name']} is now affected by {move_info['Effect']}!")

def use_item(player, pokemon):
    """Allows a player to use an item from their inventory."""
    inventory = player_inventories[player]
    if not inventory:
        print("No items available!")
        return False

    print("\nAvailable Items:")
    for i, (item, count) in enumerate(inventory.items(), start=1):
        print(f"  {i}. {item} (x{count})")

    while True:
        try:
            choice = int(input("\nChoose an item (enter number, or 0 to cancel): ")) - 1
            if choice == -1:
                return False

            item_name = list(inventory.keys())[choice]
            apply_item_effect(item_name, pokemon)
            inventory[item_name] -= 1
            if inventory[item_name] <= 0:
                del inventory[item_name]
            return True
        except (ValueError, IndexError):
            print("Invalid choice, try again.")

def apply_item_effect(item_name, pokemon):
    """Applies the effect of an item to the Pok√©mon."""
    item = items[item_name]
    if "heal" in item:
        healed_amount = min(item["heal"], pokemon_data[pokemon["name"]]["HP"] - pokemon["HP"])
        pokemon["HP"] += healed_amount
        print(f"{pokemon['name']} healed {healed_amount} HP!")
    if "remove_status" in item:
        if item["remove_status"] is True or pokemon["Status"] == item["remove_status"]:
            pokemon["Status"] = None
            print(f"{pokemon['name']} is cured of its status effect!")
    if "restore_energy" in item:
        pokemon["Energy"] += item["restore_energy"]
        print(f"{pokemon['name']} restored {item['restore_energy']} Energy!")

def ai_choose_move(pokemon):
    """AI chooses the best move based on energy and power."""
    available_moves = {m: d for m, d in pokemon["Moves"].items() if d["Energy Cost"] <= pokemon["Energy"]}
    if not available_moves:
        print(f"{pokemon['name']} has no energy to attack!")
        return None
    return max(available_moves, key=lambda m: available_moves[m]["Power"])
```

```python
while True:
    print("\nWelcome to Pok√©mon Battle!")
    show_available_pokemon()

    mode = input("\nChoose game mode: 1 for Player vs Player, 2 for Player vs AI: ").strip()
    player1_pokemon = choose_pokemon("Player 1")
    player2_pokemon = choose_pokemon("Player 2") if mode == "1" else random.choice(list(pokemon_data.keys()))

    p1 = {"name": player1_pokemon, **pokemon_data[player1_pokemon], "Energy": 100}
    p2 = {"name": player2_pokemon, **pokemon_data[player2_pokemon], "Energy": 100}

    turn = 1
    print(f"\nThe battle begins! {p1['name']} vs {p2['name']}!")

    weather = random.choice(list(weather_conditions.keys()))
    print(f"\nThe weather is {weather}!")

    while p1["HP"] > 0 and p2["HP"] > 0:
        print("\n--- Turn " + str(turn) + " ---")
        print(f"Weather: {weather}")

        current, opponent = (p1, p2) if turn % 2 == 1 else (p2, p1)
        is_ai_turn = mode == "2" and turn % 2 == 0
        player_name = "Player 1" if turn % 2 == 1 else "Player 2"
        print(f"\n‚ñ∂ {current['name']}'s turn!")

        display_status(current)

        if is_ai_turn:
            move_choice = ai_choose_move(current)
            if move_choice:
                print(f"\n{current['name']} chose {move_choice}")
                execute_move(current, opponent, move_choice, weather)
            else:
                print(f"{current['name']} is out of energy and skips turn!")
        else:
            while True:
                action = input("\nChoose action: (1) Attack, (2) Use Item: ").strip()
                if action == "1":
                    move_list = list(current["Moves"].keys())
                    while True:
                        try:
                            move_index = int(input("\nChoose a move (1-4): ")) - 1
                            if 0 <= move_index < len(move_list):
                                move_choice = move_list[move_index]
                                break
                            else:
                                print("Invalid choice, try again.")
                        except ValueError:
                            print("Invalid input, enter a number between 1-4.")
                    execute_move(current, opponent, move_choice, weather)
                    break
                elif action == "2":
                    item_used = use_item(player_name, current)
                    if item_used:
                        break
                else:
                    print("Invalid choice, enter 1 or 2.")

        if opponent["HP"] <= 0:
            print(f"\nüèÜ {current['name']} wins!")
            break

        turn += 1

    restart = input("\nRestart the game? (Y/N): ").strip().lower()
    if restart != "y":
        print("\nThanks for playing Pok√©mon Battle!")
        break
```
